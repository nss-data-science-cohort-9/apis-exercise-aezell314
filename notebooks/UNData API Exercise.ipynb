##### Question 1

Use the API to get all available data for the GDP per capita, PPP (constant 2021 international $) indicator. Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.

```python
# Set an initial endpoint to see how many items are returned from our query
endpoint_initial='https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?format=json'
response = requests.get(endpoint_initial)
total_rows=response.json()[0]['total']

# Set a final endpoint that inclues a per_page query string to ensure we retrieve all available rows 
endpoint_final=f'{endpoint_initial}&per_page={total_rows}'
response = requests.get(endpoint_final)

gdp_df=pd.json_normalize(response.json()[1])
gdp_df
```

##### Bonus Question 5

If you haven't already done so and you would like to get some additional practice using loops, use the page parameter in order to pull all records. Do not change the value of the per_page parameter. You will likely need to utilize a loop of some kind in order to pull all records.

```python
# Set an initial endpoint to see how many pages are returned from our query
endpoint_initial='https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD;SP.DYN.LE00.IN?source=2&format=json'
response = requests.get(endpoint_initial)
total_pages=response.json()[0]['pages']
```

```python
# Initialize an empty list to store our results
full_results=[]

# Loop through pages to ensure we retrieve all available rows 
for i in range(1,total_pages+1):
    endpoint_page=f'{endpoint_initial}&page={i}'
    response = requests.get(endpoint_page)
    # Add the results from the current page to the full_results list
    full_results.append(response.json()[1])
```

```python
# Convert the list of results to a DataFrame

# Start by loading just the first page into a DataFrame
full_results_df=pd.json_normalize(full_results[0])

# Then, iterate over the remaining pages and append each one to the DataFrame
for i in range(1,len(full_results)):
    full_results_df=pd.concat([full_results_df, pd.json_normalize(full_results[i])])

full_results_df
```
