```python
import requests
import pandas as pd
```

##### Question 1

Use the API to get all available data for the GDP per capita, PPP (constant 2021 international $) indicator. Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.

```python
# Set an initial endpoint to see how many items are returned from our query
endpoint_initial='https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?format=json'
response = requests.get(endpoint_initial)
total_rows=response.json()[0]['total']

# Set a final endpoint that inclues a per_page query string to ensure we retrieve all available rows 
endpoint_final=f'{endpoint_initial}&per_page={total_rows}'
response = requests.get(endpoint_final)

gdp_df=pd.json_normalize(response.json()[1])
gdp_df
```

##### Question 2

Now, use the API to get all available data for Life expectancy at birth, total (years). This indicator has code "SP.DYN.LE00.IN". Again, convert the results to a DataFrame.

```python
# Set an initial endpoint to see how many items are returned from our query
endpoint_initial='https://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN?format=json'
response = requests.get(endpoint_initial)
total_rows=response.json()[0]['total']

# Set a final endpoint that inclues a per_page query string to ensure we retrieve all available rows 
endpoint_final=f'{endpoint_initial}&per_page={total_rows}'
response = requests.get(endpoint_final)

le_df=pd.json_normalize(response.json()[1])
le_df
```

##### Question 3

Merge the two results DataFrames together. You may want to rename or drop columns prior to merging.

```python
gdp_le_df=pd.merge(left = gdp_df[['date','value','country.value','countryiso3code']].rename(columns={'value': 'GDP_Per_Capita', 'date': 'Year', 'country.value': 'Country','countryiso3code':'Country_Code'}), 
                right = le_df[['date','value','country.value','countryiso3code']].rename(columns={'value': 'Life_Expectancy', 'date': 'Year', 'country.value': 'Country','countryiso3code':'Country_Code'}), 
                how = 'inner')
```

```python
gdp_le_df
```

##### Question 4

You can also get more information about the available countries (region, capital city, income level classification, etc.) by using the Country API. Use this API to pull in all available data. Merge this with your other datasets. Use this to now remove the rows that correspond to regions and not countries.

```python
# Set an initial endpoint to see how many items are returned from our query
endpoint_initial='https://api.worldbank.org/v2/country/all?format=json'
response = requests.get(endpoint_initial)
total_rows=response.json()[0]['total']

endpoint_final=f'{endpoint_initial}&per_page={total_rows}'
response = requests.get(endpoint_final)

country_df=pd.json_normalize(response.json()[1])
country_df
```

```python
gdp_le_df=pd.merge(left = gdp_le_df, 
                   right = country_df.rename(columns={'id':'Country_Code'}), 
                   how = 'inner')
# Dropping regions
gdp_le_df=gdp_le_df[gdp_le_df['region.value'] != 'Aggregates']
gdp_le_df
```
